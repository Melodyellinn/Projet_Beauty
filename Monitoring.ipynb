```python
import pandas as pd
import sqlite3 as sql 
from datetime import datetime,timedelta
from sklearn import datasets

from evidently.test_suite import TestSuite
from evidently.test_preset import DataStabilityTestPreset

from evidently.report import Report
from evidently.metric_preset import DataDriftPreset

import smtplib as smt

iris_data = datasets.load_iris(as_frame='auto')
iris_frame = iris_data.frame
```

```python
conn = sql.connect('Application_prod.db')
train = pd.read_sql("SELECT * from Raw_data",conn)
test = pd.read_sql("SELECT * from Test_data",conn)
train["date"] = pd.to_datetime(train["date"])
test["date"] = pd.to_datetime(train["date"])
```

```python
referance_date = "20220930"
today_date = datetime.strptime(referance_date,"%Y%m%d")
referance_date_2 = "20220915"
last_date = datetime.strptime(referance_date_2,"%Y%m%d")
reference = train.copy()[train['date'].between(referance_date_2,referance_date)]
```
